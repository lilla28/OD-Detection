```python
import math

import numpy as np
import scipy.ndimage as ndi
import skimage.exposure
import skimage.morphology
from skimage import io
from skimage import measure
from skimage.filters import thresholding
from skimage.transform import resize
import pywt

from skimage.color import rgb2gray
from matplotlib import pyplot as plt

import time
```

```python
def show_img(img):
    skimage.io.imshow(img)
    skimage.io.show()
    
def arr_comp_sum(arr):
    comp_w = comp_h = 2
    shape = (arr.shape[0] // comp_h, comp_h ,arr.shape[1] // comp_w, comp_w)
    return arr.reshape(shape).sum(-1).sum(1) / 2
```

```python
img = skimage.io.imread('C:/Users/LaraLiL/Desktop/image090.jpg')  # change to parameter
print(np.info(img))
show_img(img)
```

```python
# Select the most dominant color channel(Red)
R = img[:, :, 0]
show_img(R)

# Median filtering (7x7)
ROI = ndi.median_filter(R, (7, 7))
show_img(ROI)

# Otsu thresholding
level = skimage.filters.threshold_otsu(ROI)

# Creating binary ROI mask
ROI = ROI > (level / 5)
#show_img(ROI)

# Binary erosion (contracting ROI)
for _ in range(3):
    ROI = skimage.morphology.binary_erosion(ROI)
```

```python
# Finding ROI diameter - first and last column with white pixel
temp = np.where(ROI > 0)
first_col = np.min(temp[1])
last_col = np.max(temp[1])

# Diameter estimate
D = last_col - first_col

# Zoom
zoom = 630 / D

# Reshape Img, ROI
height, width = img.shape[:2]
newIm = skimage.transform.resize(img, (int(height * zoom), int(width * zoom)))

newROI = skimage.transform.resize(ROI, (int(height * zoom), int(width * zoom)))

# Returns
print(f'ROI First col: {first_col}, ROI Last col: {last_col}')
print(f'ZOOM: {zoom}')
print(f'Original Image high, width: {img.shape[:2]}, New Image high, width: {newIm.shape[:2]}')
print(f'Original ROI high, width: {ROI.shape[:2]}, New ROI high, width: {newROI.shape[:2]}')
show_img(img)
show_img(newIm)
show_img(newROI)
```

```python
pyrnewIm = skimage.exposure.equalize_adapthist(newIm[:,:,1])
pyrnewIm = skimage.transform.resize(pyrnewIm, (512, 512))

show_img(newIm[:,:,1])
show_img(pyrnewIm)
```

```python
comp_im = np.copy(pyrnewIm)
seged = np.copy(pyrnewIm)

for _ in range(5):
    comp_im = arr_comp_sum(comp_im)
    show_img(comp_im)

thr = np.mean(comp_im)
comp_im[comp_im < thr] = 0
show_img(comp_im)

seged[np.repeat(np.repeat(comp_im, 2**5, axis = 1), 2**5, axis = 0) == 0] = 0


show_img(seged)
```

```python
pyrnewROI = skimage.transform.resize(newROI, (512,512))
show_img(pyrnewROI)
```

```python
SE_kor = skimage.morphology.selem.disk(45, np.uint64)
pyrnewROI = skimage.morphology.erosion(pyrnewROI, SE_kor)
show_img(pyrnewROI)
```

```python
seged[pyrnewROI == 0] = 0
show_img(seged)
```

```python

SE = skimage.morphology.disk(5,np.uint8)

J = skimage.morphology.dilation(pyrnewIm,SE)
show_img(J)
```

```python
# VERSION 2
seged2 = np.zeros((2 * 512, 2 * 512))
SE_kor = skimage.morphology.disk(45, np.uint8)
seged2[512][512] = 1
seged2 = skimage.morphology.dilation(seged2, SE_kor)

show_img(seged2)
```

```python
SumIm = J.copy()

# first col, first row 
np.cumsum(SumIm[:, 0], out = SumIm[:, 0])
np.cumsum(SumIm[0, :], out = SumIm[0, :])

# rest
for k in range(1, SumIm.shape[0]):
    for l in range(1, SumIm.shape[1]):
        SumIm[k][l] += SumIm[k - 1][l] + SumIm[k][l - 1] - SumIm[k - 1][l - 1]


# Info
print(SumIm[511, 511])
print(np.sum(J))
print(np.average(SumIm))

show_img(SumIm)
```

```python
mask = np.copy(seged)
show_img(seged2)
show_img(SumIm)
show_img(mask)
show_img(J)

wind_size = 54
kor_ter = 6553
negyzet_ter = (2 * wind_size + 1)**2

probMap = np.zeros((512, 512), dtype = np.float32)

for i in range(1 + wind_size, 512 - wind_size):
    for j in range(1 + wind_size, 512 - wind_size):
        if (mask[i][j] != 0):
            seged3 = seged2[512-i:512-i+512, 512-j:512-j+512]
            
            # Circle
            sum_under_circle = np.sum(J[seged3 == 1])
            
            # Rectange
            top_left = SumIm[i - 1 - wind_size][j - 1 - wind_size]
            top_rght = SumIm[i - 1 - wind_size][j + wind_size]
            btm_left = SumIm[i + wind_size][j - 1 - wind_size]
            btm_rght = SumIm[i + wind_size][j + wind_size]
            sum_under_rect = btm_rght - btm_left - top_rght + top_left
            
            # Probability
            probMap[i][j] = ((sum_under_circle / kor_ter) / ((sum_under_rect - sum_under_circle) / (negyzet_ter - kor_ter)))
            
show_img(probMap)
```

```python
probMap[probMap < 1.25] = 0
show_img(probMap)
```

```python
result = skimage.transform.resize(probMap, (630,D))
# Scale to 0-255
img_data = 255 * result / result.max()
# Converting to uint8
img_out = img_data.astype(np.uint8)

show_img(img_out)
```
